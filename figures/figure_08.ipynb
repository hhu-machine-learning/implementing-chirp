```python
import os
os.chdir('..')

import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import oifits as oi
import vlbi
import vlbi_utils as utils
from skimage.feature import register_translation
from scipy.ndimage import fourier_shift
from skimage.measure import compare_ssim
```

```python
pathlist = [['./data/syntheticData/staticBlackhole/UVdata/blackhole-36-10.oifits',
             './data/syntheticData/staticBlackhole/targetImgs/blackhole-36-10.fits'],
            ['./data/syntheticData/staticBlackhole/UVdata/blackhole-54-10.oifits',
             './data/syntheticData/staticBlackhole/targetImgs/blackhole-54-10.fits']]
```

```python
%matplotlib notebook
for path in pathlist:
    name = path[0].split('/')[-1].split('.')[0]
    
    oidata = oi.open(path[0], quiet=True)
    fitsdata = fits.open(path[1])[0]
    src_im = fitsdata.data
    fov = np.abs(fitsdata.header['CDELT1']) * np.pi/180 * fitsdata.header['NAXIS1']
    naxis = fitsdata.header['NAXIS1']
    
    chirp = vlbi.CHIRP(oidata, fov, naxis)
    res = chirp.reconstruct(display=True)
    chirp.saveFits('./results/' + name + '.fits')
```

```python
def score(res_im, ref_im):

    shift = register_translation(ref_im, res_im)[0]
    shifted_res = fourier_shift(np.fft.fft2(res_im), shift)
    shifted_res = np.real(np.fft.ifft2(shifted_res))

    mse = np.linalg.norm(shifted_res - ref_im)
    drange = np.max(shifted_res) - np.min(shifted_res)
    ssim = compare_ssim(ref_im, shifted_res, data_range=drange)

    return mse, ssim
```

```python
reslist = ['./results/blackhole-36-10.fits',
            './results/blackhole-54-10.fits']
fitslist = ['./data/syntheticData/staticBlackhole/targetImgs/blackhole-36-10.fits',
            './data/syntheticData/staticBlackhole/targetImgs/blackhole-54-10.fits']
targetlist = ['./data/syntheticData/targets/blackhole-36.pgm',
              './data/syntheticData/targets/blackhole-54.pgm']
```

```python
%matplotlib inline
fig, axs = plt.subplots(2, 3, figsize=(9,6.6))
plt.subplots_adjust(wspace=-0.1)
s = 0.9
for i in range(2):
    
    fitsdata = fits.open(fitslist[i])[0]
    fov = np.abs(fitsdata.header['CDELT1']) * np.pi/180 * fitsdata.header['NAXIS1']
    src_im = fitsdata.data
    target_im = plt.imread(targetlist[i])[:,:,0]
    target_im = target_im.astype(np.float64)
    target_im /= np.max(target_im)
    target_im *= np.max(src_im)
    res = fits.open(reslist[i])[0].data
    src_im = src_im.flatten()
    src_im = np.reshape(src_im,fitsdata.data.shape,order='F')[::-1,:]
    res = res.flatten()
    res = np.reshape(res,fitsdata.data.shape,order='F')[::-1,:] 
    mse, ssim = score(res, target_im)
    
    minVal = np.min([src_im, target_im, res])
    maxVal = np.max([src_im, target_im, res])
    im = axs[i,0].imshow(src_im, cmap='gray', vmin=minVal, vmax=maxVal)
    temp = fig.colorbar(im, ax=axs[i,0], shrink=s, label='Jy/pixel')
    temp.remove()
    fov_muas = fov * 1e6 * 3600 * 180 / np.pi
    ticklabels = np.linspace(fov_muas/2, -fov_muas/2, 7, dtype=int)
    ticks = np.linspace(0, src_im.shape[0]-1, 7)
    axs[i,0].set_xticks(ticks)
    axs[i,0].set_xticklabels([])
    axs[i,0].set_yticks(ticks)
    axs[i,0].set_yticklabels(ticklabels)
    axs[i,0].set_ylabel('Declination [$\mu$as]')
    im = axs[i,1].imshow(target_im, cmap='gray', vmin=minVal, vmax=maxVal)
    temp = fig.colorbar(im, ax=axs[i,1], shrink=s, label='Jy/pixel')
    temp.remove()
    axs[i,1].set_xticks(ticks)
    axs[i,1].set_xticklabels([])
    axs[i,1].set_yticks(ticks)
    axs[i,1].set_yticklabels([])
    im = axs[i,2].imshow(res, cmap='gray', vmin=minVal, vmax=maxVal)
    fig.colorbar(im, ax=axs[i,2], shrink=s, label='Jy/pixel')
    axs[i,2].set_xticks(ticks)
    axs[i,2].set_xticklabels([])
    axs[i,2].set_xlabel('MSE: %0.6f\nSSIM: %0.4f' %(mse, ssim))
    axs[i,2].set_yticks(ticks)
    axs[i,2].set_yticklabels([])
axs[0,0].set_title('Source')
axs[0,1].set_title('Target')
axs[0,2].set_title('Reconstruction')
axs[-1,0].set_xticklabels(ticklabels)
axs[-1,0].set_xlabel('Right Ascension [$\mu$as]')
axs[-1,1].set_xticklabels(ticklabels)
axs[-1,1].set_xlabel('Right Ascension [$\mu$as]')
axs[-1,2].set_xticklabels(ticklabels)
axs[-1,2].set_xlabel('Right Ascension [$\mu$as]\nMSE: %0.6f\nSSIM: %0.4f' %(mse, ssim))
```
