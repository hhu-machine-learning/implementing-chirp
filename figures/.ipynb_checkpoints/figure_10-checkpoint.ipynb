```python
import os
os.chdir('..')

import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import oifits as oi
import vlbi
import vlbi_utils as utils
from skimage.feature import register_translation
from scipy.ndimage import fourier_shift
from skimage.measure import compare_ssim
```

```python
pathlist = [['./data/syntheticData/staticCelestial/UVdata/celestial-14-10.oifits',
             './data/syntheticData/staticCelestial/targetImgs/celestial-14-10.fits'],
            ['./data/syntheticData/staticNatural/UVdata/natural-03-10.oifits',
             './data/syntheticData/staticNatural/targetImgs/natural-03-10.fits']]
lams = [0.1, 0.01, 0.001, 0.0001, 0.00001]
```

```python
%matplotlib notebook
for path in pathlist:
    name = path[0].split('/')[-1].split('.')[0]

    oidata = oi.open(path[0], quiet=True)
    fitsdata = fits.open(path[1])[0]
    src_im = fitsdata.data
    fov = np.abs(fitsdata.header['CDELT1']) * np.pi/180 * fitsdata.header['NAXIS1']
    naxis = fitsdata.header['NAXIS1']

    chirp = vlbi.CHIRP(oidata, fov, naxis)
    
    for lam in lams:
        res = chirp.reconstruct(lam=lam, display=True)
        chirp.saveFits('./results/' + name + '_lam_' + str(lam) + '.fits')
```

```python
def score(res_im, ref_im):

    shift = register_translation(ref_im, res_im)[0]
    shifted_res = fourier_shift(np.fft.fft2(res_im), shift)
    shifted_res = np.real(np.fft.ifft2(shifted_res))

    mse = np.linalg.norm(shifted_res - ref_im)
    drange = np.max(shifted_res) - np.min(shifted_res)
    ssim = compare_ssim(ref_im, shifted_res, data_range=drange)

    return mse, ssim
```

```python
%matplotlib inline
fig, axs = plt.subplots(2,5, figsize=(10,4.5))

src_im1 = fits.open('./data/syntheticData/staticCelestial/targetImgs/celestial-14-10.fits')[0].data
target_im1 = plt.imread('./data/syntheticData/targets/celestial-14.pgm')[:,:,0]
target_im1 = target_im1.astype(np.float64)
target_im1 /= np.max(target_im1)
target_im1 *= np.max(src_im1)

src_im2 = fits.open('./data/syntheticData/staticNatural/targetImgs/natural-03-10.fits')[0].data
target_im2 = plt.imread('./data/syntheticData/targets/natural-03.pgm')[:,:,0]
target_im2 = target_im2.astype(np.float64)
target_im2 /= np.max(target_im2)
target_im2 *= np.max(src_im2)

for j, lam in enumerate(lams):
    axs[0,j].set_title('$\lambda$ = ' + str(lam))

for j, lam in enumerate(lams):
    res1 = fits.open('./results/celestial-14-10_lam_' + str(lam) + '.fits')[0].data
    mse1, ssim1 = score(res1, target_im1)
    axs[0,j].imshow(res1, cmap='gray')
    axs[0,j].set_xticks([])
    axs[0,j].set_xlabel('MSE: %0.6f\nSSIM: %0.4f' %(mse1, ssim1))
    axs[0,j].set_yticks([])
    res2 = fits.open('./results/natural-03-10_lam_' + str(lam) + '.fits')[0].data
    mse2, ssim2 = score(res2, target_im2)
    axs[1,j].imshow(res2, cmap='gray')  
    axs[1,j].set_xticks([])
    axs[1,j].set_xlabel('MSE: %0.6f\nSSIM: %0.4f' %(mse2, ssim2))
    axs[1,j].set_yticks([])
```
