```python
import os
os.chdir('..')

import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
targetlist = ['./results/paper/target',
              './results/paper/clean',
              './results/paper/squeeze',
              './results/paper/bsmem',
              './results/paper/chirp']
fitslist = ['./data/syntheticData/staticBlackhole/targetImgs/blackhole-36-10.fits',
            './data/syntheticData/staticBlackhole/targetImgs/blackhole-54-10.fits',
            './data/syntheticData/staticCelestial/targetImgs/celestial-03-10.fits',
            './data/syntheticData/staticCelestial/targetImgs/celestial-09-10.fits',
            './data/syntheticData/staticCelestial/targetImgs/celestial-14-10.fits',
            './data/syntheticData/staticCelestial/targetImgs/celestial-01-10.fits',
            './data/syntheticData/staticNatural/targetImgs/natural-03-10.fits']
```

```python
fig, axs = plt.subplots(6, 7, figsize=(10,8.5))
for j in range(7):
    im = fits.open(fitslist[j])[0].data
    if(j == 0 or j == 1):
        imf = im.flatten()
        im = np.reshape(imf,im.shape,order='F')[::-1,:] 
    axs[0,j].imshow(im, cmap='gray')
    axs[0,j].set_xticks([])
    axs[0,j].set_yticks([])
for i in range(1,6):
    for j in range(7):
        im = plt.imread(targetlist[i-1] + str(j + 1) + '.pgm')
        axs[i,j].imshow(im, cmap='gray')
        axs[i,j].set_xticks([])
        axs[i,j].set_yticks([])

axs[0,0].set_title('A')
axs[0,1].set_title('B')
axs[0,2].set_title('C')
axs[0,3].set_title('D')
axs[0,4].set_title('E')
axs[0,5].set_title('F')
axs[0,6].set_title('G')

axs[0,0].set_ylabel('Source')
axs[1,0].set_ylabel('Target')
axs[2,0].set_ylabel('CLEAN')
axs[3,0].set_ylabel('SQUEEZE')
axs[4,0].set_ylabel('BSMEM')
axs[5,0].set_ylabel('CHIRP')
```

```python
fig.savefig("./figure_07.pdf",  bbox_inches="tight")
```
