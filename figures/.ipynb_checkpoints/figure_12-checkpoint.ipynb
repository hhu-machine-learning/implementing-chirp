```python
%matplotlib inline
fig, axs = plt.subplots(2, 3, figsize=(10,5))
s = 1.02

for i, d in enumerate(data):
    fitsdata = fits.open('./data/realData/origBUData/' + d +'.IMAP')
    ref = fitsdata[0].data[0,0][::-1,:]
    fov = fitsdata[0].header['NAXIS1'] * np.abs(fitsdata[0].header['CDELT1']) * np.pi/180
    fov_mas = fov * 1e3 * 3600 * 180 / np.pi
    ticklabels = np.linspace(fov_mas/2, -fov_mas/2, 7, dtype=int)
    ticks = np.linspace(0, ref.shape[0]-1, 7)
    res = fits.open('results/realData/' + d + '.fits')[0].data
    
    im1 = axs[0,i].imshow(ref, cmap='gray')
    im2 = axs[1,i].imshow(res, cmap='gray')
    if(i == 2):
        fig.colorbar(im1, ax=axs[0,i], shrink=s, label='Jy/pixel')
        fig.colorbar(im2, ax=axs[1,i], shrink=s, label='Jy/pixel')
    else:
        fig.colorbar(im1, ax=axs[0,i], shrink=s)
        fig.colorbar(im2, ax=axs[1,i], shrink=s)
    axs[0,i].set_xticks(ticks)
    axs[0,i].set_xticklabels([])
    axs[0,i].set_yticks(ticks)
    axs[0,i].set_yticklabels([])
    axs[1,i].set_xticks(ticks)
    axs[1,i].set_xticklabels(ticklabels)
    axs[1,i].set_xlabel('Right Ascension [mas]')
    axs[1,i].set_yticks(ticks)
    axs[1,i].set_yticklabels([])

axs[0,0].set_title('3C111')
axs[0,1].set_title('3C446')
axs[0,2].set_title('OJ287')
axs[0,0].set_yticklabels(ticklabels)
axs[0,0].set_ylabel('BU Results\nDeclination [mas]')
axs[1,0].set_yticklabels(ticklabels)
axs[1,0].set_ylabel('Our Results\nDeclination [mas]')
```
