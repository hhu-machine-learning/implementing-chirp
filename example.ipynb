```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import oifits as oi
import vlbi
import vlbi_utils as utils
```

# Black Hole

```python
flux = 1.0
fluxstr = str(flux).split('.')[0] + str(flux).split('.')[1]
%matplotlib inline
fig, axs = plt.subplots(2,3, figsize=(10,6.66), sharex=True, sharey=True)
plt.subplots_adjust(hspace=0.3)
for c, val in enumerate(['01', '18', '36', '54', '72', '90']):
    fitsdata = fits.open('./data/syntheticData/staticBlackhole/targetImgs/blackhole-' + val + '-' + fluxstr + '.fits')
    axs[c//3,c%3].imshow(fitsdata[0].data, cmap='gray')
    axs[c//3,c%3].set_title(val)
```

```python
val = '90'
flux = 1.0
fluxstr = str(flux).split('.')[0] + str(flux).split('.')[1]
oidata = oi.open('./data/syntheticData/staticBlackhole/UVdata/blackhole-' + val + '-' + fluxstr + '.oifits')
fitsdata = fits.open('./data/syntheticData/staticBlackhole/targetImgs/blackhole-' + val + '-' + fluxstr + '.fits')[0]
im = fitsdata.data
fov = np.abs(fitsdata.header['CDELT1']) * np.pi/180 * fitsdata.header['NAXIS1']
naxis = fitsdata.header['NAXIS1']
```

```python
chirp = vlbi.CHIRP(oidata, fov, naxis)
```

```python
%matplotlib notebook
res = chirp.reconstruct(display=True)
```

```python
%matplotlib inline
chirp.showResult()
```
